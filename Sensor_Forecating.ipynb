```python
def list_to_3d_array(data, step_size):
    segments = []
    for sublist in data:
        for i in range(len(sublist) - step_size + 1):
            segment = sublist[i:i + step_size]
            segments.append(segment)

    # Convert to numpy array and reshape to (n, step_size, 1)
    segments_array = np.array(segments)
    segments_3d = segments_array.reshape(-1, step_size, 1)

    return segments_3d
```

```python
def split_3d_array(array, size1):
    n, step_size, _ = array.shape
    size2 = step_size - size1

    # Split the array along the second dimension
    array1 = array[:, :size1, :]
    array2 = array[:, size1:, :]

    return array1, array2
```
